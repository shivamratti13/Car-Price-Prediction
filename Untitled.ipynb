```python
df_train = pd.get_dummies(df_train, columns=['brand','model_year','accident','fuel_type'], dtype= np.uint8)
```

```python
df_train.columns
```

```python
df_train.head()
```

```python
column_to_move = 'price'

# Pop the column and store it
column = df_train.pop(column_to_move)

# Insert the column at the end
df_train[column_to_move] = column
```

```python
df_train.head()
```

```python
df_train_sub = df_train.drop(columns=['id', 'model', 'engine',
       'transmission', 'ext_col', 'int_col', 'clean_title',
       ], inplace=False)
```

```python
df_train_sub.head()
```

```python
from sklearn.preprocessing import StandardScaler
column_to_scale = 'milage'

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the selected column
df_train_sub[column_to_scale] = scaler.fit_transform(df_train_sub[[column_to_scale]])
```

```python
df_train_sub.head()
```

```python
X_train = df_train_sub.to_numpy()[:,:-1]
X_train
```

```python
y_train = df_train_sub.to_numpy()[:,-1]
y_train
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
df_train_sub['linear_predictions'] = model.predict(X_train)
```

```python
df_train_sub.head()
```

```python
from sklearn.metrics import root_mean_squared_error
rmse = root_mean_squared_error(df_train_sub['price'], df_train_sub['linear_predictions'])
rmse
```

## Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
```

```python
X_poly = poly.fit_transform(X_train)
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_poly, y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
df_train_sub['polynomial_prediction'] = reg.predict(X_poly)
```

```python
df_train_sub.tail()
```

```python
from sklearn.metrics import root_mean_squared_error
rmse = root_mean_squared_error(df_train_sub['price'], df_train_sub['polynomial_prediction'])
rmse
```
